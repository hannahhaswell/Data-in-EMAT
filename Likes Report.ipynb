### Import Pandas

```python
import pandas as pd
```

The data below shows the most recent accounts I have liked media on and the time I did so.

```python
mlikes_df = pd.DataFrame(likes['media_likes'], columns = ['timestamp', 'account'])
mlikes_df
```

The data below shows the most recent accounts I have liked comments on and the time I did so.

```python
clikes_df = pd.DataFrame(likes['comment_likes'], columns = ['timestamp', 'account'])
clikes_df
```

Below is data on the media I have liked sorted by timestamp.

```python
mgroup_df = mlikes_df.groupby('account').count().sort_values('timestamp', ascending = False)
mgroup_df
```

Below is data on the comments I have liked sorted by timestamp.

```python
cgroup_df = clikes_df.groupby('account').count().sort_values('timestamp', ascending = False)
cgroup_df
```

### Import pyplot for a Bar Graph and a Pie Graph of my Instagram Data

```python
from matplotlib import pyplot as plt
```

### Below is a bar graph and a pie plot of my Instagram media likes.

```python
sortcount = mgroup_df.sort_values('timestamp' ,ascending=False)
labels = sortcount['timestamp'].keys()
vals = sortcount['timestamp']
width = .75

fig, ax = plt.subplots(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
ax.bar(labels, vals, width)
ax.set_ylabel('Likes')
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

The results of the bar graph above are hard to read, but give me a broad sense of data analysis of my likes on media on Instagram. The broad analysis is that I like media fewer times for more accounts.

```python
fig2, ax2 = plt.subplots()
ax2.pie(vals, labels = labels)
ax2.axis('equal')
plt.show()
```

The results of the pie graph above show that I have liked many different accounts media. It seems that I like media fewer times on more accounts compared to just liking a lot of media for a few particular accounts.  

### Below is a bar graph and a pie plot of my Instagram comment likes.

```python
sortcount = cgroup_df.sort_values('timestamp' ,ascending=False)
labels = sortcount['timestamp'].keys()
vals = sortcount['timestamp']
width = .75

fig, ax = plt.subplots(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
ax.bar(labels, vals, width)
ax.set_ylabel('Likes')
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

The results of the bar graph above show that I have liked the user nicolebarnes25 comments the most. This graph could give us more information if it was formatted better, but for the purpose of this report Dr. Silva says that this format is fine.

```python
fig2, ax2 = plt.subplots()
ax2.pie(vals, labels = labels)
ax2.axis('equal')
plt.show()
```

The results of this pie graph above show that I have liked the user nicolebarnes25 comments the most and then paigeeee99, melinalynn_, and emandem63 seem to be a close tie of how many of their comments I have liked. It also shows that I have liked very few comments on a lot of accounts.
